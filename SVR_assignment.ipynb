```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
dataset = pd.read_csv('Car_Purchasing_Data.csv', encoding = 'Latin-1')
x = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

```
y = y.reshape(len(y),1)
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
y_train = sc_y.fit_transform(y_train)
y_test = sc_y.transform(y_test)
```

```
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(x_train,y_train)
```

```
y_pred = sc_y.inverse_transform(regressor.predict(x_test))
x_test = sc_x.inverse_transform(x_test)
y_test = sc_y.inverse_transform(y_test)
```

```
plt.scatter(x_test[:,1],y_test, color='red')
plt.scatter(x_test[:,1],y_pred, color='blue')
plt.xlabel('Age')
plt.ylabel('Car Purchase Amount')
plt.show()
```

```
plt.scatter(x_test[:,2],y_test, color='red')
plt.scatter(x_test[:,2],y_pred, color='blue')
plt.xlabel('Annual Salary')
plt.ylabel('Car Purchase Amount')
plt.show()
```

```
plt.scatter(x_test[:,3],y_test, color='red')
plt.scatter(x_test[:,3],y_pred, color='blue')
plt.xlabel('Credit Card Debt')
plt.ylabel('Car Purchase Amount')
plt.show()
```
